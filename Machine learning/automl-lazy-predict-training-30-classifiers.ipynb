```python
# Get the Data
data = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/train.csv')
```

```python
# Shape of Data
print(data.shape)
```

```python
# Number of Unique Target Class
print(data.target.unique())
```

```python
# Encoding Target Variable
target_dict = {'Class_1' : 0,
               'Class_2' : 1,
               'Class_3' : 2,
               'Class_4' : 3}
data['target'] = data['target'].map(target_dict)
```

```python
#Checking Class Imbalance
print(data.target.value_counts())
```

```python
# Checking Missing Values
print(data.isnull().sum())
```

### Lazy Predict - https://lazypredict.readthedocs.io/en/latest/

```python
# !pip install lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
# Separate features and target
X = data.iloc[:,1:-1]
y= data.target

# Split Train and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.20,random_state =123)
```

```python
# LazyClassifier object
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
model_dictionary = clf.provide_models(X_train,X_test,y_train,y_test)
print(models)
```
